# get the baseline

```python
def get_word_label(line):
    s=line.split()

    word=s[1]
    label=s[4]
    return word, label


def get_word_label_dict(file):
    word_label_dict={}
    with open(file,'r') as fin:
        lines = fin.readlines()
        for line in lines:
            if len(line)>1:
                if line[0]!='#':
                    word,label=get_word_label(line)
                    if word not in word_label_dict:
                        word_label_dict[word]={}
                    if label not in word_label_dict[word]:
                        word_label_dict[word][label]=0                

                    word_label_dict[word][label]+=1
    return word_label_dict

def get_the_most_frequent_label(d):
    s = sorted(d.items(),key=lambda item:item[1])
    return s[-1][0]


def get_most_frequent_word_label_dict(word_label_dict):
    most_frequent_word_label_dict={}
    for word in word_label_dict:
        label = get_the_most_frequent_label(word_label_dict[word])
        most_frequent_word_label_dict[word]=label
    return most_frequent_word_label_dict

def evaluate(test_dict,most_dict):
    token_sum=0
    correct_sum=0
    for word in test_dict:
        for label in test_dict[word]:
            token_sum+=test_dict[word][label]
            if word in most_dict:
                if label ==most_dict[word]:
                    correct_sum+=test_dict[word][label]
    print('acc=',correct_sum/token_sum)
    return token_sum,correct_sum,correct_sum/token_sum
```

```python
# 1 read train data
train_file = 'example/data/en_ewt-ud/en_ewt-ud-train.conllu'
word_label_dict = get_word_label_dict(train_file)
most_frequent_word_label_dict=get_most_frequent_word_label_dict(word_label_dict)
```

```python
# 2 read dev/test data
test_file='example/data/en_ewt-ud/en_ewt-ud-test.conllu'
test_word_label_dict = get_word_label_dict(test_file)
dev_file='example/data/en_ewt-ud/en_ewt-ud-dev.conllu'
dev_word_label_dict = get_word_label_dict(dev_file)
```

```python
# 3.1 test set, baseline acc
evaluate(test_word_label_dict,most_frequent_word_label_dict)
```

```python
# 3.2 dev set, baseline acc
evaluate(dev_word_label_dict,most_frequent_word_label_dict)
```

```python
# 3.3 train set, baseline acc
evaluate(word_label_dict,most_frequent_word_label_dict)
```
